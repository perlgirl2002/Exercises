```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
columns = ['time', 'username', 'computer', 'process_name', 'start_or_stop']
df = pd.read_csv("C:\\Users\\kim\\data\\proc.txt", nrows=500000)
df.columns=columns
```

```python
df.head()
```

```python
# Time Plot
time = pd.DataFrame(df.time.value_counts()).reset_index()
time.columns=(['time','count'])
time = time.head(25)
time.head()
```

```python
# One variable 4 different ways:
sns.distplot(time['count'])
```

```python
sns.distplot(time['count'], kde=False, rug=True)
```

```python
sns.kdeplot(time['count'], shade=True, cut=0)
```

```python
from scipy import stats
sns.distplot(time['count'], kde=False, fit=stats.gamma);
```

```python
# Two variables 3 different ways:
# ONE
sns.barplot(x='time', y='count', data = time)
```

```python
# TWO
sns.lmplot(x='time', y='count', data=time, fit_reg=False)
```

```python
# THREE
sns.pairplot(time)
```

```python
# One continuous and one categorical plot 6 ways
df = df.sample(200)
sns.stripplot(x="time", y="start_or_stop", data=df, jitter=True);
```

```python
sns.swarmplot(x="time", y="start_or_stop", data=df);
```

```python
sns.boxplot(x="time", y="start_or_stop", data=df);
```

```python
sns.violinplot(x="time", y="start_or_stop", data=df);
```

```python
sns.violinplot(x="time", y="start_or_stop", data=df, inner=None)
sns.swarmplot(x="time", y="start_or_stop", data=df, color="w", alpha=.5);
```

```python
sns.countplot(x="start_or_stop", data=df, palette="Greens_d");
```
